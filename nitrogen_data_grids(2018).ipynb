```python
import numpy as np
import pandas as pd
```

```python
filename = '2018年全球人工氮肥施用量(grids).csv'
df = pd.read_csv(filename, index_col='grid_num')
```

```python
# 将求和结果添加为新列到原DataFrame中
df['fertilizer_input'] = df.sum(axis=1)
df = df.reset_index()
```

```python
df = df.merge(pd.DataFrame({'grid_num':range(1, 360*720+1)}), on='grid_num', how='right')
df.head()
```
